```python
import pandas as pd
import numpy as np
```

```python
# pass in column names for each CSV
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv('u.user', sep='|', names=u_cols,
                    encoding='latin-1')

r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('u.data', sep='\t', names=r_cols,
                      encoding='latin-1')

# the movies file contains columns indicating the movie's genres
# let's only load the first five columns of the file with usecols
m_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url']
movies = pd.read_csv('u.item', sep='|', names=m_cols, usecols=range(5),
                     encoding='latin-1')
```

```python
#Merge datasets
movie_ratings = pd.merge(movies, ratings)  #merge is nothing but inner join
lens = pd.merge(movie_ratings, users)
```

```python
lens.head()
```

```python
#Let us answer some simple questions now. 
#Most rated movies
 #How many users have rated for each movie
most_rated = lens.groupby('title').size()  #returns a tuple  
most_rated.head(10)  
```

```python
lens.title.value_counts()[:20]     #does same thing as group_by , in addition it sorts too in descending order
```

**Important**

```python
#Highest rated movies  #Aggregation
highest_ratings = lens.groupby('title').agg({'rating':[np.size,np.mean]})
highest_ratings.head(10)
```

```python
highest_ratings.sort([('rating', 'mean')], ascending=False).head() #not useful becoz no. of ratings not taken into account
```

```python
atleast_100 = highest_ratings['rating']['size'] >= 250
highest_ratings[atleast_100].sort([('rating', 'mean')], ascending=False)[:20]  #sorted by mean with size grater than 250
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(users.age,bins=10)
plt.title("Distribution of users' ages")
plt.ylabel('count of users')
plt.xlabel('age');
```

```python
#dividing data into specific buckets
#grouping now on the basis of age_group
labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
lens['age_group'] = pd.cut(lens.age, range(0, 81, 10), right=False, labels=labels)
lens['age_group'].head(10)
lens.head(5)
```

```python
lens.groupby('age_group').agg({'rating': [np.size, np.mean]})
```

Note : the means above are close but comparable
Observations:
1.Younger people rate more
2.Older people watch better rated/non-cheesy movies hence the higher mean rating in that age group

```python
most_100 = lens.groupby('movie_id').size()[:100]
```

```python
lens.set_index('movie_id', inplace=True)
```

```python
by_age = lens.loc[most_100.index].groupby(['title', 'age_group'])
by_age.rating.mean().head(15)


#Interprtation:
#1. Ace:Ventura movie is liked by elderly people more etc..    
```

```python
by_age.rating.mean().unstack(1).fillna(0)[10:20]
```

```python
lens.reset_index('movie_id', inplace=True)
```

```python
print type(lens)
#pd.pivot?
pivoted = lens.pivot_table(index=['movie_id', 'title'],  #pivot first does group_by and then pivot on basis of columns
                           columns=['sex'],
                           values='rating',
                           fill_value=0)
pivoted.head()
```

```python
pivoted['diff'] = pivoted.M - pivoted.F
pivoted.head()
```

```python
pivoted.reset_index('movie_id', inplace=True)
```

```python
disagreements = pivoted[pivoted.movie_id.isin(most_100.index)]['diff']
disagreements.head(10)
```

```python
disagreements.describe()
```

```python
disagree = disagreements.copy()
disagree.sort()
disagree.head()
```

```python
#Horizontal bar graph

disagree.plot(kind='barh', figsize=[9, 15])
plt.title('Male vs. Female Avg. Ratings\n(Difference > 0 = Favored by Men)')
plt.ylabel('Title')
plt.xlabel('Average Rating Difference');
```
